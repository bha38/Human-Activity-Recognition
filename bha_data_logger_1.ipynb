```python
# Her dosyaya yazdrta tarih yazdrlmal.

import serial
import csv

ser_port = "/dev/cu.usbserial-AL01GV88"
baud_rate = 38400 # Set baud rate

# Open the serial port
ser = serial.Serial(ser_port,baud_rate,timeout=1) # timeout ???

ser.flushInput()

while True:
    try:
        ser_byte = ser.readline().strip()
        str_byte = str(ser_byte)
        
        new_str=str_byte[2:-1]
        new_list=new_str.split('\\t')
        print(new_list)
        
        with open("sensor_log.csv","a") as f:
            writer = csv.writer(f,delimiter=",")
            writer.writerow(new_list)
            
    except:
        print("Keyboard Interrupt")
        ser.close()
        break
```
